```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [ i for i in range(100)]
x = np.array(x)
x = 2*np.pi * x* 0.01
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,color='red', marker = 'o', label='Sin')
plt.plot(x,z,color='green', marker = 's', label='Cos')
plt.title('Sin Plot')
plt.xlabel('x')
plt.ylabel('sin')
plt.show()
```
